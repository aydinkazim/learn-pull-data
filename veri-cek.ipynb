```python
import requests
from bs4 import BeautifulSoup
url = "http://www.tureb.org.tr/tr/RehberVeritabani/AraOda?odaId=21"

r = requests.get(url)

soup = BeautifulSoup(r.content,"html.parser")
```

```python
gelen_veri = soup.find_all("table",{"class":"display"})
```

```python
gelen_veri
```

## Erişmek istedğimiz veri < tbody > elemanının içinde olduğu için onun indexine göre veri çekeceğiz.

```python
rehberler = gelen_veri[0].contents[3]
```

```python
rehberler
```

```python
rehberler_tr = rehberler.find_all("tr")
```

```python
rehberler_tr
```

```python
for rehber in rehberler_tr:
    rehber_bilgileri = rehber.find_all("td")
    rehber_isimleri = rehber_bilgileri[0].text
    rehber_odasi = rehber_bilgileri[1].text
    print(rehber_isimleri,"Odası:",rehber_odasi)
```
